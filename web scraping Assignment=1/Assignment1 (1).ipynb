
# 2) Write a python program to display IMDB’s Top rated 50 movies’ data (i.e. name, rating, year of release)

```python
url='https://www.imdb.com/chart/top/'
```

```python
IMBD = requests.get(url)
```

```python
soup=BeautifulSoup(IMBD.content)
```

```python
movie_name=[i.text.split('\n')[2]for i in soup.find_all(class_="titleColumn")]
movie_name
```

```python
rating=[i.text.strip()for i in soup.find_all(class_="ratingColumn imdbRating")]
rating
```

```python
year=[i.text.strip() for i in soup.find_all(class_='secondaryInfo')]
year
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({"Name":movie_name,'Rating':rating,'year':year})
df
```

Write a python program to scrape mentioned details from dineout.co.in and make data framei) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
restaurants_name=[i.text.strip()for i in soup.find_all(class_='restnt-name ellipsis')]
restaurants_name
```

```python
restaurants_Cuisine=[i.text.strip()for i in soup.find_all(class_='double-line-ellipsis')]
restaurants_Cuisine
```

```python
Location = [i.text.strip()for i in soup.find_all(class_='restnt-loc ellipsis')]
Location
```

```python
Ratings= [i.text.strip()for i in soup.find_all(class_='restnt-rating rating-4')]
Ratings
```

```python
images = []
for i in soup.find_all('img',class_='no-img'):
    images.append(i.get('data-src'))
images    
```

```python
df = pd.DataFrame({'restaurants name':restaurants_name,'restaurants Cuisine':restaurants_Cuisine,'Location':Location,'Ratings':Ratings})
df
```
